```python
# finding the sum of each products in 2015
item_total1 = Agric_data.groupby('Item')['Y2015'].sum()
item_total1
```

```python
# finding the sum of each products in 2018
item_total2 = Agric_data.groupby('Item')['Y2018'].sum()
item_total2
```

```python
# Select 'Y2017' and 'Area' columns
area_data = Agric_data[['Y2017', 'Area']]
area_data
```

```python
# Group by 'Area' and calculate the sum
area_totals = area_data.groupby('Area').sum()
area_totals
```

```python
# Find the area with the highest sum in 2017
highest_area = area_totals['Y2017'].idxmax()
highest_area
```

```python
# Highest sum of the highest area
highest_sum = Agric_data.groupby('Area')['Y2017'].sum().max()
highest_sum
```

```python
# finding the area 7th lowest area
lowest_7th_area = area_totals['Y2017'].nsmallest(7).index[-1]
lowest_7th_area
```
